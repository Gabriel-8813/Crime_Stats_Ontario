```python
import pandas as pd
```

## ETL for 1st dataframe " Crimes_Stats_ON"

```python
# Crime file path

crimes_path = "../Crime_Stats_Ontario/Resources/Crimes_Stats_ON.csv"  
crimes_Stats = pd.read_csv(crimes_path)
crimes_Stats.head()
```

```python
crimes_Stats.info()
```

```python
#Dropping unecessary columns

Dropped = ['STATUS','SYMBOL','TERMINATED','DECIMALS']
crimes_Stats = crimes_Stats.drop(columns=Dropped)
```

```python
# Split the columns GEO into 2 new columns Povinces and Cities
crimes_Stats[['Cities', 'Provinces']] = crimes_Stats['GEO'].str.split(',', expand=True)
crimes_Stats.head()
```

```python
#Diplay updated Dataframe
crimes_Stats.head()
```

```python
crimes_Stats = crimes_Stats.drop(columns=['GEO'])
```

```python
crimes_Stats[['Violation', 'Violation_ID']] = crimes_Stats['Violations'].str.split('[', expand=True)
crimes_Stats.head()
```

```python
crimes_Stats.head()
```

```python
crimes_Stats = crimes_Stats.drop(columns=['Violations'])
```

```python
crimes_Stats.head()
```

```python
crimes_Stats[['Province', 'City_ID']] = crimes_Stats['Provinces'].str.split('[', expand=True)
```

```python
crimes_Stats
```

```python
crimes_Stats['Violation_ID'] = crimes_Stats['Violation_ID'].str.replace(']', '', regex=False)
```

```python
crimes_Stats
```

```python

Dropped = ['STATUS','SYMBOL','TERMINATED','DECIMALS']
crimes_Stats = crimes_Stats.drop(columns=Dropped)

Col_to_clean = ['Violation_ID', 'City_ID']
```

```python
Col_to_clean 
```

```python
for col in Col_to_clean:
    crimes_Stats[col] = crimes_Stats[col].str.replace(']', '', regex=False)
```

```python
crimes_Stats
```

```python
crimes_Stats = crimes_Stats.drop(columns=['UOM', 'UOM_ID'])
crimes_Stats
```
