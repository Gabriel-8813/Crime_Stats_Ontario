```python
#Import Dependencides

import pandas as pd
```

## ETL for 1st dataframe " Crimes_Stats_ON"

```python
# Import Crimes Data from csv file, and store in a Dataframe

crimes_path = "Resources\Crimes_Stats_ON.csv"
crimes_Stats = pd.read_csv(crimes_path)
crimes_Stats.head()
```

```python
# Display the dataframe info

crimes_Stats.info()
```

```python
# Split the column GEO into 2 new columns Povinces and Cities
crimes_Stats[['Cities', 'Provinces']] = crimes_Stats['GEO'].str.split(',', expand=True)


# Get the Violations ID and the violations description into 2 separate columns 'Violation' and 'Violation_ID'
crimes_Stats[['Violation', 'Violation_ID']] = crimes_Stats['Violations'].str.split('[', expand=True)


# Split the column Provinces into 2 new columns'Province' and 'City_ID'
crimes_Stats[['Province', 'City_ID']] = crimes_Stats['Provinces'].str.split('[', expand=True)


crimes_Stats.head()
```

```python
#Drop the unecessary columns

Dropped = ['DGUID', 'VECTOR', 'GEO', 'COORDINATE', 'Violations', 'Provinces', 'Violation_ID', 'Province', 'City_ID', 'STATUS','SYMBOL','TERMINATED','DECIMALS', 'SCALAR_FACTOR', 'SCALAR_ID', 'UOM', 'UOM_ID']
crimes_Stats1 = crimes_Stats.drop(columns=Dropped)
```

```python
# Chreck the updated df
crimes_Stats1
```

```python
#Delete the 2013 entries in the df, to match the dataset in Hate Crimes csv

crimes_Stats1 = crimes_Stats1[crimes_Stats1['REF_DATE'] != 2013]
```

```python
# Replace NaN values with a 0 

crimes_Stats1['VALUE'] = crimes_Stats1['VALUE'].fillna(0)
```

```python
# Change the values display in VALUE column, If it's a whole number, show it without .00. Otherwise, keep the float format

def custom_format(x):
    if x == int(x):  
        return int(x)
    else:            
        return x

# Apply formatting 
crimes_Stats1['VALUE'] = crimes_Stats1['VALUE'].apply(custom_format)
```

```python
# Reorder columns

crimes_Stats1 = crimes_Stats1[['REF_DATE', 'Cities', 'Statistics', 'Violation', 'VALUE']]
crimes_Stats1
```

```python
crimes_Stats1['Statistics'] = crimes_Stats1['Statistics'].str.replace('Rate,', 'Rate', regex=False)
crimes_Stats1['Statistics'] = crimes_Stats1['Statistics'].str.replace('Total,', 'Total', regex=False)
```

```python
crimes_Stats1['Statistics'] = crimes_Stats1['Statistics'].str.replace(r'\bPercentage\b', 'Percent', regex=True)
```

```python
# Extract the first word and create a new column

crimes_Stats1['Statistics_Groups'] = crimes_Stats1['Statistics'].str.split().str[0]

# Group by the first word
groups = crimes_Stats1.groupby('Statistics_Groups')

```

```python
crimes_Stats1
```

```python
statistics_df = crimes_Stats1.copy()

statistics_df = statistics_df.rename(columns={ 
    'Cities': 'City', 
    'REF_DATE': 'Year', 
    'Statistics': 'Statistics_detail', 
    'Violation': 'Violation', 
    'VALUE': 'Value', 
    'Statistics_Groups': 'Statistic_Type' 
})

statistics_df
```

```python
Statistics = "Statistics.csv"
statistics_df.to_csv(Statistics, index=True, index_label='Statistic_ID')
```

```python
statistics_df.info()
```

```python
metropolitan_cities_df = crimes_Stats1.groupby("Cities").first().reset_index()
metropolitan_cities_df = metropolitan_cities_df[['Cities']].rename(columns={'Cities': 'City'})

metropolitan_cities_df
```

```python
Cities = "Cities.csv"
metropolitan_cities_df.to_csv(Cities, index=True, index_label='City_ID')
```

```python
metropolitan_cities_df.info()
```

```python
Years_df = crimes_Stats1.groupby("REF_DATE").first().reset_index()
Years_df = Years_df[['REF_DATE']].rename(columns={'REF_DATE': 'Year'})

Years_df
```

```python
Years = "Years.csv"
Years_df.to_csv(Years, index=True, index_label='Year_ID')
```

```python
Years_df.info()
```

```python
statistic_types_df = crimes_Stats1.groupby("Statistics_Groups").first().reset_index()
statistic_types_df = statistic_types_df[['Statistics_Groups']].rename(columns={'Statistics_Groups': 'Statistic_Type'})
statistic_types_df
```

```python
Statistic_Types = "Statistic_Types.csv"
statistic_types_df.to_csv(Statistic_Types, index=True, index_label='Stat_Type_ID')
```

```python
statistic_types_df.info()
```

```python
Violations_df = crimes_Stats1.groupby("Violation").first().reset_index()
Violations_df = Violations_df[['Violation']]
Violations_df
```

```python
Violations = "Violations.csv"
Violations_df.to_csv(Violations, index=True, index_label='Violation_ID')
```

```python
Violations_df.info()
```
