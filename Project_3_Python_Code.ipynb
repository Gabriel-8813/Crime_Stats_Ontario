# Dataset 1 - Total Violent Criminal code violations (excluding traffic violation) 
## Transforming the data into user friendly dataframes to be stored in SQLite Database.

```python
# Import Dependencies
import pandas as pd
```

```python
# Import Crimes Data from csv file, and store in a Dataframe

crimes_path = "Resources/Crimes_Stats_ON.csv"  
crimes_Stats_old = pd.read_csv(crimes_path)
crimes_Stats_old.head()
```

```python
# Display the dataframe info

crimes_Stats_old.info()
```

```python
#Drop unnecessary columns

Dropped = ['STATUS','SYMBOL','TERMINATED','DECIMALS']
crimes_Stats1 = crimes_Stats_old.drop(columns=Dropped)
```

```python
# Split the columns GEO into 2 new columns Provinces and Cities

crimes_Stats1[['Cities', 'Provinces']] = crimes_Stats1['GEO'].str.split(',', expand=True)
crimes_Stats1.head()
```

```python
#Display updated Dataframe

crimes_Stats1.head()
```

```python
# Drop unnecessary column GEO

crimes_Stats1 = crimes_Stats1.drop(columns=['GEO'])
```

```python
# Split the Violations ID and the violations description in 2 separate columns 'Violation' and 'Violation_ID'

crimes_Stats1[['Violation', 'Violation_ID']] = crimes_Stats1['Violations'].str.split('[', expand=True)
crimes_Stats1.head()
```

```python
# Drop unnecessary column Violations

crimes_Stats1 = crimes_Stats1.drop(columns=['Violations'])
```

```python
# Chreck the updated df

crimes_Stats1.head()
```

```python
# Split the column Provinces into 2 new columns'Province' and 'City_ID'

crimes_Stats1[['Province', 'City_ID']] = crimes_Stats1['Provinces'].str.split('[', expand=True)
```

```python
# Chreck the updated df

crimes_Stats1
```

```python
# Get the Violations ID and the violations description into 2 separate columns 'Violation' and 'Violation_ID'

crimes_Stats1['Violation_ID'] = crimes_Stats1['Violation_ID'].str.replace(']', '', regex=False)
```

```python
# Chreck the updated df

crimes_Stats1
```

```python
# Iterates over specified columns to remove ']' from string values for each column in the df

Col_to_clean = ['Violation_ID', 'City_ID']
for col in Col_to_clean:
    crimes_Stats1[col] = crimes_Stats1[col].str.replace(']', '', regex=False)
```

```python
# Check the changes

crimes_Stats1.head()
```

```python
#Drop unnecessary columns in the Crimes statistics data

Drop_crime_col = ['SCALAR_ID','DGUID','Provinces']
crimes_Stats_new = crimes_Stats1.drop(columns=Drop_crime_col)
crimes_Stats_new.head()
```

```python
# Rename the column header from REF_DATE to Year

crimes_Stats_new = crimes_Stats_new.rename(columns={'REF_DATE': 'Year'})
```

```python
# Rearrange the columns in the data frame to be more useful for reading the dataset

crime_cols = ['Year', 'Province','Cities','City_ID','Violation','Violation_ID', 
            'Statistics', 'UOM', 'UOM_ID', 'VALUE', 'SCALAR_FACTOR', 'VECTOR','COORDINATE']
crimes_Stats_new = crimes_Stats_new[crime_cols]
crimes_Stats_new
```

```python
# Split into DataFrames that represent the different type of Units of Measurement (UOM) 

Crime_stats_num = crimes_Stats_new[crimes_Stats_new['UOM'] == 'Number']
Crime_stats_per = crimes_Stats_new[crimes_Stats_new['UOM'] == 'Percent']
Crime_stats_rate = crimes_Stats_new[crimes_Stats_new['UOM'] == 'Rate']
```

```python
# Display the crime statistics with the UOM in numbers representing every incident in absolute number count

Crime_stats_num
```

```python
# Drop rows where 'Violations' column contains the word 'Total' as these represent sub totals in the dataset

Crime_stats_df = Crime_stats_num[~Crime_stats_num['Violation'].str.contains('Total', case=True, na=False)]
Crime_stats_df
```

```python
# create a column with unique values to enable analysis
# Combine values from 'Year' and 'Cities' columns to create a new unique column

Crime_stats_df.loc[:, 'location_year'] = Crime_stats_df['Year'].astype(str) + '_' + Crime_stats_df['Cities'].astype(str)
```

```python
# Check the df changes 

Crime_stats_df
```

```python
# Create a dataframe from the existing dataframe to be used for analysis
# Selecting specific columns to create a new DataFrame

location_year_df = Crime_stats_df[['Year','Province','Cities']]
location_year_df
```

```python
# Create a dataframe from the existing dataframe to be used for analysis
# Selecting specific columns to create a new DataFrame

CrimeStats1 = Crime_stats_df[['Year','Cities', 'Violation', 'Violation_ID', 'VALUE', 'location_year']]
CrimeStats1
```

```python
# Rename the following columns to be more meaning to data analysis 'VALUE_x' to 'Crime_incidents', 'Type of motivation' to 'Hate Crime',  'VALUE_y' to 'Hate_crime_count'

CrimeStats1 = CrimeStats1.rename(columns={'VALUE': 'Crime_incidents'})
CrimeStats1
```

```python
# Display crime statistics with UOM in percentages representing the 
# year-over-year (current year over last year) percentage change in the rate of actual incidents

Crime_stats_per.head()
```

```python
# Display crime statistics with the UOM in rate representing total persons charged per 100,000 population of the sub group within the population

Crime_stats_rate.head()
```

```python
#Dropping unnecessary columns in the Crimes statistics data

Dropped_crime_col = ['UOM','UOM_ID','SCALAR_FACTOR', 'COORDINATE','VECTOR']
Crime_stats_rate = Crime_stats_rate.drop(columns=Dropped_crime_col)
```

```python
# Rename the column header from REF_DATE to Year

Crime_stats_rate = Crime_stats_rate.rename(columns={'VALUE': 'Crime_rate'})
```

```python
# Check df updates

Crime_stats_rate
```

```python
# Drop rows where 'Violations' column contains the word 'Total' as these represent sub totals in the dataset

CrimeStats_rate_df = Crime_stats_rate[~Crime_stats_rate['Violation'].str.contains('Total', case=True, na=False)]
CrimeStats_rate_df
```

```python
# Identify NaN objects in the column 'value' and replace with Zero (0)

Crime_stats_rate_nan = Crime_stats_rate.isna().any()
Crime_stats_rate_nan
```

```python
# Store the final Dataframe into a csv file to be used to create the database

CrimeStats_rate_df.to_csv('CrimeStats_rate output.csv', index=True, header=True)
```

# Dataset 2 - Police-reported hate crime, by type of motivation, selected regions and Canada (selected police services)
Transform the data to be most useful for an analysis of Hate Crime by type of motivation in cities in Ontario. 
Modify the statistics to be able to merge and analysis with the Crime Statistics by violation type  

```python
# Import Crimes Data from csv file, and store in a Dataframe

motivations_path = "Resources/Hate_Crimes_Stats_ON.csv"  
motivations_df = pd.read_csv(motivations_path)
motivations_df 
```

```python
# Get df info

motivations_df.info()
```

```python
#Drop unnecessary columns in the Hate Crimes statistics

Drop_col = ['STATUS','SYMBOL','TERMINATED','DECIMALS']
hate_crime = motivations_df.drop(columns=Drop_col)
```

```python
# Check the change

hate_crime.head()
```

```python
# Split the columns GEO into 2 new columns Provinces and Cities

hate_crime[['location', 'details']] = hate_crime['GEO'].str.split('[', expand=True)
hate_crime
```

```python
# Iterates over specified columns to remove ']' from string values for each column in hate_crime

hc_to_clean = ['details', 'location']
for col in hc_to_clean:
    hate_crime[col] = hate_crime[col].str.replace(']', '', regex=False)
```

```python
# Check updates

hate_crime
```

```python
# Drop extra columns

hate_crime = hate_crime.drop(columns=['GEO', 'UOM', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID'])
hate_crime
```

```python
# Display current columns
hate_crime.columns
```

```python
# Create another data frame with selected columns to hold the Canada, selected police services

new_columns = ['REF_DATE', 'DGUID', 'Type of motivation', 'VECTOR', 'COORDINATE', 'VALUE', 'location', 'details']
Hate_crime_CAN = hate_crime[new_columns]
Hate_crime_CAN
```

```python
# Replace all NaN values in the 'VALUE' column with zero

Hate_crime_CAN['VALUE'] = Hate_crime_CAN['VALUE'].fillna(0)
Hate_crime_CAN
```

```python
#Rename the 'REF_DATE' to 'Year', dropping the DGUID, coordinate and details columns

Hate_CAN_final = Hate_crime_CAN.rename(columns={'REF_DATE': 'Year', 'Type of motivation':'Hate_Crime', 'VALUE': 'Hate_crime_count'})
```

```python
# Create a dataframe from the existing dataframe to be used for analysis
# Selecting specific columns to create a new DataFrame 'HateStats'

HateStats = Hate_CAN_final[['Year','Hate_Crime', 'Hate_crime_count', 'location']]
HateStats
```

```python
# Drop rows where 'location' is 'Ontario'

Hate_crime_CAN2 = Hate_crime_CAN[Hate_crime_CAN['location'] != 'Ontario ']
Hate_crime_CAN2
```

```python
# Drop rows where 'location' is 'Canada, selected police services
# 	'
Hate_crime_CAN3 = Hate_crime_CAN[Hate_crime_CAN['location'] != 'Canada, selected police services']
Hate_crime_CAN3.head(50)
```

```python
#Rename the 'REF_DATE' to 'Year', dropping the DGUID, coordinate and details columns

Hate_crime_CAN_final = Hate_crime_CAN3.rename(columns={'REF_DATE': 'Year'})
Hate_crime_CAN_final = Hate_crime_CAN_final.drop(columns=['DGUID', 'COORDINATE', 'details'])
Hate_crime_CAN_final
```

```python
# Merge the Ontario hate crime data frame with the crime statistics data frame using the 'year' as the inner join 

crime_hate_merged_df = pd.merge(Crime_stats_num, Hate_crime_CAN_final, on='Year', how='inner')
crime_hate_merged_df
```

```python
# Drop columns in the merged df for ease of further analysis

crime_hate_merged_df1 = crime_hate_merged_df.drop(columns=['UOM', 'UOM_ID', 'SCALAR_FACTOR','COORDINATE', 'VECTOR_x', 'VECTOR_y'])
crime_hate_merged_df1
```

```python
# Rename the following columns to be more meaning to data analysis 'VALUE_x' to 'Crime_incidents', 'Type of motivation' to 'Hate Crime',  'VALUE_y' to 'Hate_crime_count'

crime_hate_merged_df2 = crime_hate_merged_df1.rename(columns={'VALUE_x': 'Crime_incidents', 'Type of motivation': 'Hate_Crime', 'VALUE_y': 'Hate_crime_count'})
crime_hate_merged_df2
```

```python
# Check the merged data frame for 'NaN' values and display

crime_hate_merged_nan = crime_hate_merged_df2[crime_hate_merged_df2.isna().any(axis=1)]
crime_hate_merged_nan
```

```python
# Replace all NaN values in the 'VALUE' column with zero

crime_hate_merged_df2['Crime_incidents'] = crime_hate_merged_df2['Crime_incidents'].fillna(0)
crime_hate_merged_df2
```

```python
# Drop rows where 'Violations' column contains the word 'Total' as these represent sub totals in the dataset

crime_hate_merged_df3 = crime_hate_merged_df2[~crime_hate_merged_df2['Violation'].str.contains('Total', case=True, na=False)]
crime_hate_merged_df3
```

```python
# Save the merged data into a csv file to load into SQLITE

crime_hate_merged_df3.to_csv('Crime&Hate stat output.csv', index=True, header=True)
```
